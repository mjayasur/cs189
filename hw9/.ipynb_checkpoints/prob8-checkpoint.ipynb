# Classifying and Clustering Cell Images

```python
import pandas as pd
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
col_names = list(data.feature_names)
col_names.append('target')
df = pd.DataFrame(np.concatenate((data.data, data.target), axis=1), columns=col_names)
```

```python
df
```

```python
data = df[df.columns.difference(['diagnosis', 'Unnamed: 32'])]
data
```

```python
labels = df[['id', 'diagnosis']]
labels
```

```python
data, labels = np.array(data), np.array(labels)
```

```python
np.linalg.svd(data)[1]
```

```python
pca = PCA(n_components=3)
```

```python
pca.fit(data)
```

```python
pca.singular_values_
```

```python
data_pca = pca.transform(data)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data_pca[:,0],data_pca[:,1],c=labels,cmap='plasma')
```
